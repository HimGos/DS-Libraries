```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

> #### You can use ANY library for your work.

```python
# WOrking with inbuilt dataset of seaborn

df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.plot()                    # Plot is inbuilt of Pandas

# Here it is considering only numertical columns

# on x-axis it's showing index, on y-axis it's showing values
```

```python
df.plot(kind='area')
```

> #####  alpha parameter controls the transparency of graph

```python
df.plot(kind='area', alpha=0.6)       
```

> #####  Stack is showing whole data on single scale by overlapping values

```python
df.plot(kind='area', alpha=0.6, stacked=False) 
```

##### 

```python
df.head()
```

```python
# Finding relation b/w columns....

df.plot.scatter(x='sepal_length', y='sepal_width')
```

```python
df.plot.scatter(x='sepal_length', y='petal_length')


# Here we can see if sepal length increase then petal length too
```

> ***'c' means color scale which makes a relation b/w the color column and other columns on graph. This third parameter becomes a criteria to give color to a specific dot.*** 

```python
df.plot.scatter(x='sepal_length', y='petal_length', c='sepal_width')


# We can see sepal width is used here to define the color of dot of sepal & petal length
```

> ***'s' means size which increases the size of dots or any other type of chart we are using.***

```python
df.plot.scatter(x='sepal_length', y='petal_length', c='sepal_width', s=60)
```

```python
# We can vary a size to make it dynamic by using a column name in s parameter.

df.plot.scatter(x='sepal_length', y='petal_length', c='sepal_width', s=df['sepal_length']*20)
```

##### 

```python
df.plot.hexbin(x='sepal_length', y='petal_length')
```

```python
df.plot.hexbin(x='sepal_length', y='petal_length', gridsize=10)
```

# ***3D Plots*** 

```python
from mpl_toolkits import mplot3d
```

```python
# We need 3 axis for 3D graphs

x = np.linspace(-1, 6, 30)
x
```

```python
y = np.linspace(-1, 6 , 30)
```

```python
z = x+y
```

```python
z
```

```python
# First we create axis

ax = plt.axes(projection = '3d')         # Creates grid
ax.plot3D(x,y,z)                         # Assigning coordinates

# Now we can set labels using axes

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

##### 

```python
# First we create axis

ax = plt.axes(projection = '3d')                                              # Creates grid
ax.plot3D(df['sepal_length'], df['sepal_width'], df['petal_length'])         # Assigning coordinates

# Now we can set labels using axes

ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length')
```

> ## ***Using Seaborn***

```python
sns.relplot(x='total_bill', y='tip', data=df1)

# ALL GRAPHS ARE SAME
```

```python
# Finding how many people are smokers and how many are not.

df1['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip', data=df1, hue='smoker')         # HUE categorized data with colors
```

```python
# Finding at which time more tips were given

sns.relplot(x='total_bill', y='tip', data=df1, hue='time')
```

```python
# Finding who gave more tip

sns.relplot(x='total_bill', y='tip', data=df1, hue='sex')
```

> ### Taking 2 columns as parameter

```python
sns.relplot(x='total_bill', y='tip', data=df1, hue='time', style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=df1, hue='time', style='sex')
```
