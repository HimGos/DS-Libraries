# Grouping

```python
# Grouby function involves some combination of ->  
"""  

    splitting the object   +  Apply Function  +  Combine Results

"""
```

#### Value count of each countries to see countries that have the most results for this particular survey.

```python
df['Country'].value_counts()
```

##### 

```python
# STEP 1 :-  SPLIT THE OBJECT

# Group all of the results by country

country_group = df.groupby(['Country'])                              
```

```python
country_group.get_group('United States').head()         # We get dataframe with countries as USA only
```

#### 

```python
# STEP 2 :- APPLY FUNCTION

# To get result of single country, we can use loc
filt = df['Country'] == 'India'
df.loc[filt]['SocialMedia'].value_counts()
```

```python
# To get result of multiple countries, we use groupby

country_group['SocialMedia'].value_counts().loc['India']
```

```python
# Getting percentage of above data

country_group['SocialMedia'].value_counts(normalize=True).loc['India']
```

```python
# Getting median salaries of all countries

country_group['ConvertedComp'].median()
```

```python
# Getting median salaries of specific countries

country_group['ConvertedComp'].median().loc['Germany']
```

##### If you wanna run multiple aggregate functions on your group, use agg

```python
# Let say I wanna see both median and mean 

country_group['ConvertedComp'].agg(['median', 'mean'])
```

```python
country_group['ConvertedComp'].agg(['median', 'mean']).loc['Canada']
```

##### Let say I wanna find out how many people in each country know how to use python

```python
# To find for single country we can use simple filter

filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# To find for multiple countries

country_group['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_group['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())    # x representing whole series here
```

#### 

##### What % of people from each coountry know Python?

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
country_uses_python
```

```python
# Now we have total number of people and those who knew python in those countries
```

```python
# We will use CONCAT function to combine data

# STEP 3:-  COMBINE RESULTS

python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)

python_df
```

```python
python_df.rename(columns={'LanguageWorkedWith': 'NumKnowsPython', 'Country':'NumberOfRespondents'}, inplace=True)

python_df.head(4)
```

```python
python_df['PercKnowsPython'] = (python_df['NumKnowsPython']/ python_df['NumberOfRespondents']) *100

python_df
```

```python
python_df.sort_values(by='PercKnowsPython', ascending=False, inplace=True)

python_df.head(30)
```
