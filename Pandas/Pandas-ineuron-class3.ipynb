# MERGE

```python
data4 = {
    'emp_id': [101,102,103,104],
    'salary': [234,457,123,342],
    'pf': [23,34,54,75]
}
```

```python
data5 ={
    'emp_id': [101,102,104,105],
    'mob_no': [23423,354564,23412,1322],
    'house_no': [345,453,234,654]
}
```

```python
df4 = pd.DataFrame(data4)
```

```python
df5 = pd.DataFrame(data5)
```

```python
pd.merge(df4,df5)                  # Only mutual rows are shown
```

##### Merge by default find the common ones since inner merge is performed by default
##### ---------------------------------------------------------

```python
# LEFT MERGE

pd.merge(df4,df5, how='left')
```

```python
# RIGHT MERGE

pd.merge(df4,df5, how='right')
```

```python
# OUTER MERGE

pd.merge(df4,df5, how='outer')
```

##### Outer Merge shows everything! Fills empty fields with NaN value
##### --------------------------------------------------

```python
pd.merge(df4,df5, how='right', on='emp_id')            # ON allows us to specify which column to use for merging
```

```python
data8 = {
    'emp_id': [101,102,103,104],
    'salary': [234,457,123,342],
    'pf': [23,34,54,75]
}

data9 ={
    'emp_id': [101,102,104,105],
    'salary': [234,457,123,300],
    'house_no': [345,453,234,654]
}
```

```python
df8 = pd.DataFrame(data8)
df9 = pd.DataFrame(data9)
```

##### Merging on the basis of 2 same columns

```python
pd.merge(df8,df9, on=['emp_id', 'salary'])           # When both the columns are same, it shows those values only.
```

##### Let say one data has more number of rows

```python
data9 ={
    'emp_id': [101,102,104,105, 106],
    'salary': [234,457,123,300, 400],
    'house_no': [345,453,234,654, 342]
}
```

```python
df9 = pd.DataFrame(data9)
```

```python
pd.merge(df8, df9, on= ['emp_id','salary'])         # It doesn't matter if no. of rows is uneven
```

# JOIN

```python
data10 = {
    'emp_id1': [101,102,103,104],
    'salary1': [234,457,123,342],
    'pf': [23,34,54,75]
}

data11 ={
    'emp_id2': [101,102,104,105,106],
    'salary2': [234,457,123,300,265],
    'house_no': [345,453,234,654,111]
}
```

```python
df10 = pd.DataFrame(data10, index=['a','b','c','d'])
df11 = pd.DataFrame(data11, index=['a', 'b', 'd', 'e', 'f'])
```

```python
df10.join(df11)          # LEFT join by default
```

##### JOIN trying to perform match on the basis of INDEXES

##### MERGE try to perform match on COLUMNS

```python
df10.join(df11, how='right')
```

```python
df_sales = pd.read_csv('D:/sales_data_final.csv')
```

```python
df_sales.head(1)
```

### Task: Create a Profit_Flag column which writes positive if we are + and vice versa

```python
filt = df_sales['profit'].apply(lambda x: 'positive' if x>=0 else 'negative')
```

```python
df_sales['flag_profit'] = filt
```

```python
df_sales.head()
```

### Create column to store len of customer name

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)

df_sales.head(2)
```

### Categorize quantity into three segments - High, Med, Low

```python
filt = df_sales['quantity'].apply(lambda x : 'Low' if x<3 else 'Med' if x>3 and x<7 else 'High')
```

```python
df_sales['flag_quantity'] = filt

df_sales.head()
```

```python
df_sales['sq_quantity'] = df_sales['quantity'].apply(lambda x :x**2)
df_sales.head(2)
```
